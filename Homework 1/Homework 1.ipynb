```python
from sklearn import utils
from sklearn.datasets import load_digits
import numpy as np
import matplotlib.pyplot as plt
import import_ipynb
from Functions import shuffle, training, cce_loss
from MLP import MLP

data = load_digits(return_X_y = True)
''''
plt.imshow(data[0][0].reshape(8,8), cmap='gray')
plt.show()
'''
normalizedData = np.empty(shape = [1797,64])
targets = np.zeros(shape = [1797,10])
for i in range(1797):
    normalizedData[i,:] = data[0][i,:]/np.linalg.norm(data[0][i,:])
    j = data[1][i] 
    targets[i,j] = 1

minibatches, targets_batched = shuffle(normalizedData, targets, 100)
net = MLP(num_units_per_Layer = [32,10], size_input = 64)

losses = training(10, minibatches, targets_batched, net) 

    
```

```python
import matplotlib.pyplot as plt

pl = [item for sublist in losses for item in sublist]
t = np.linspace(0, 18000, 18000)

fig, ax = plt.subplots()
ax.plot(t, pl)

ax.set(xlabel='epochs', ylabel='loss', xlim=(0, 100), ylim = (0,5),
       title='Loss evaluation')
fig.set_figwidth(15) 
fig.set_figheight(5) 
plt.show()
```
