```python
import numpy as np
import import_ipynb
from MLP import MLP


def array_gen (many):
    x = np.random.rand(100)
    t = np.empty(100)
    for n in range(100):
        t[n] = x[n]**3 - x[n]**2
    return x,t
x,t = array_gen (100)

def mean_square(y,t):
    return 1/2*(y-t)**2
    
       
net = MLP(num_units_per_Layer = [10,1], size_input = 1)

def training(epoch,data,net):
    losses = []
    for e in range(epoch):
        for d in range(len(data)):
            losses.append(mean_square(net.forward_step(data[d]), t[d]))
            net.backward_step(losses[d])
    return losses
losses = training(1000, x, net)         
```

```python
import matplotlib.pyplot as plt

t = np.linspace(0, 100000, 100000)

fig, ax = plt.subplots()
ax.plot(t, losses)

ax.set(xlabel='epochs', ylabel='loss', xlim=(0, 1000),
       title='Loss evaluation')
fig.set_figwidth(15) 
fig.set_figheight(5) 
plt.show()
```
